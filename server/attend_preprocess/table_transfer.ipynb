```python
# ===================== 【超参数：只改这里】 =====================
SHEET_NAME    = "Year2"
REGIONS       = [
    "A2:G7",
    "J2:P7",
    "S2:Y7",
    "AA2:AG7",
    "AI2:AO7",
    "AQ2:AW7",
]
# =================================================================

# 打开 Excel
wb = load_workbook(EXCEL_FILE)
ws = wb[SHEET_NAME]

# 解析单元格：A4 → (4, 1)
def parse_cell(cell_str):
    col_str = re.findall(r'[A-Za-z]+', cell_str)[0]
    row_str = re.findall(r'\d+', cell_str)[0]
    return int(row_str), column_index_from_string(col_str)

# 解析区域：A2:G7 → 左上角、右下角
def parse_region(region_str):
    top_left, bottom_right = region_str.split(":")
    return top_left, bottom_right

# 最终大字典
final_result = {}

# ========== 批量处理所有区域 ==========
for region in REGIONS:    
    # 解析区域
    tl_cell, br_cell = parse_region(region)
    start_row, start_col = parse_cell(tl_cell)
    end_row, end_col = parse_cell(br_cell)

    # 每个区域的 Programme 取自自己区域的左上角（如 A2, J2, S2...）
    programme_cell = ws.cell(row=start_row, column=start_col)
    programme = str(programme_cell.value).strip() if programme_cell.value else "Unknown"

    programme = programme.replace(" ", "").replace("InternetofThings", "IoT")
    programme = programme.replace("TeleG1", "Tele-G1").replace("TeleG2", "Tele-G2")

    # 遍历每个区域的数据行：4,5,6,7行
    for row in range(4, end_row + 1):
        # ----------------------
        # 列定位：相对于区域起始列（自动偏移）
        # ----------------------
        code_col       = start_col          # Code = 区域第1列
        module_col     = start_col + 1      # Module Name
        lecturer1_col  = start_col + 2      # Lecturer1
        lecturer2_col  = start_col + 3      # Lecturer2
        class_col      = start_col + 5      # Class
        pattern_col    = start_col + 6      # Pattern

        # 读取值
        code          = ws.cell(row=row, column=code_col).value
        module_name   = ws.cell(row=row, column=module_col).value
        lecturer1     = ws.cell(row=row, column=lecturer1_col).value
        lecturer2     = ws.cell(row=row, column=lecturer2_col).value
        class_val     = ws.cell(row=row, column=class_col).value
        pattern_val   = ws.cell(row=row, column=pattern_col).value

        # 单元格坐标
        cell_index = ws.cell(row=row, column=code_col).coordinate

        # 存入字典
        if code:
            key = f"{programme}_{code}"

            final_result[key] = {
                "Code": code,
                "cell_index": cell_index,
                "Module Name": module_name,
                "Class": class_val,
                "Programme": programme,
            }

# ========== 输出最终结果 ==========
for code, info in final_result.items():
    print(f"{code}: {info}")

print(f"\n🎉 总计提取：{len(final_result)} 个模块")

with open("year2_class.json", "w") as file:
    json.dump(final_result, file, ensure_ascii=False, indent=4)
```

# 匹配

```python
def find_matched_fill_cells(EXCEL_FILE: str, SHEET_NAME: str, TARGET_CELL: str) -> list[dict]:
    """匹配与 TARGET_CELL 填充色相同的单元格，返回排序后的坐标和值（仅在 TARGET_CELL 下方）。"""
    MATCH_PATTERN_TYPE = True

    def fill_color_key(cell) -> tuple:
        """只看填充样式 cell.fill，不看字体色 cell.font"""
        fill = cell.fill
        fg = fill.fgColor
        bg = fill.bgColor

        base = (
            fg.type,
            fg.rgb,
            fg.theme,
            fg.indexed,
            fg.tint,
            bg.type,
            bg.rgb,
            bg.theme,
            bg.indexed,
            bg.tint,
        )

        return (fill.patternType, *base) if MATCH_PATTERN_TYPE else base

    def coord_sort_key(coord: str) -> tuple[int, int]:
        col_letters, row_num = coordinate_from_string(coord)
        return (column_index_from_string(col_letters), row_num)

    wb = load_workbook(EXCEL_FILE, data_only=True)
    ws = wb[SHEET_NAME]

    target_key = fill_color_key(ws[TARGET_CELL])

    # 限制匹配列范围：仅在 TARGET_CELL 右侧 1~5 列内查找
    target_col_letters, target_row = coordinate_from_string(TARGET_CELL)
    target_col_idx = column_index_from_string(target_col_letters)
    min_col = target_col_idx + 1
    max_col = min(target_col_idx + 6, ws.max_column)

    matched_cells: list[dict] = []
    for row in ws.iter_rows(min_row=target_row + 1, min_col=min_col, max_col=max_col):
        for cell in row:
            try:
                if fill_color_key(cell) == target_key:
                    matched_cells.append({
                        "cell": cell.coordinate,
                        "room": cell.value,
                    })
            except Exception:
                continue

    matched_cells_sorted = sorted(matched_cells, key=lambda x: coord_sort_key(x["cell"]))
    return matched_cells_sorted
```

# 整合成新样式

```python
def get_weekday_en(s: str) -> str:
    """传入形如 '2-Mar' 的字符串，按固定年份 2026 返回英文星期。"""
    dt = datetime.strptime(f"2026-{s}", "%Y-%d-%b")
    weekday_en = dt.strftime("%A")
    return weekday_en

def class_range_to_student_count(class_range: str) -> int:
    """输入班级范围字符串（如 '17-20', '1~5', '1～5', '17'），返回总人数。"""
    text = str(class_range).strip().replace(" ", "")
    text = text.replace("～", "~")

    # 统一提取数字；支持 17-20 / 1~5 / 17
    nums = [int(x) for x in re.findall(r"\d+", text)]
    if not nums:
        raise ValueError(f"无效班级范围: {class_range}")

    if len(nums) == 1:
        start = end = nums[0]
    else:
        start, end = nums[0], nums[1]

    if start > end:
        start, end = end, start

    def class_size(class_no: int) -> int:
        if 1 <= class_no <= 16:
            return 30  # Telecom(1-10) + IoT(11-16)
        if 17 <= class_no <= 24:
            return 25  # EIE(17-20) + IST(21-24)
        raise ValueError(f"班级号超出定义范围(1-24): {class_no}")

    return sum(class_size(i) for i in range(start, end + 1))

_PROGRAMME_CLASS_MAP = {
    "Tele-G1": "1-5",
    "Tele-G2": "6-10",
    "IoT1":    "11-13",
    "IoT2":    "14-16",
    "EIE":     "17-20",
    "IST":     "21-24",
    "IoT":     "11-16",
}

def programme_to_student_count(programme: str) -> int:
    """传入 programme 名称，返回对应总人数。"""
    class_range = _PROGRAMME_CLASS_MAP.get(str(programme).strip())
    if class_range is None:
        raise ValueError(f"未知 programme: {programme}")
    return class_range_to_student_count(class_range)
```

```python
def excel_to_dict_with_openpyxl(file_path):
    """
    纯 openpyxl 读取 Excel，生成你要的字典格式
    key: {Stream}_{Code}
    value: first_half, second_half, week1-15
    """
    wb = load_workbook(file_path, data_only=True)
    ws = wb.active  # 获取第一个工作表

    # 获取表头，定位列
    headers = {}
    for col_idx, cell in enumerate(ws[1], 1):
        headers[cell.value] = col_idx

    # 你需要的列
    col_stream = headers["Stream"]
    col_code = headers["Code"]
    col_first = headers["First-half"]
    col_second = headers["Second-half"]

    # 周数列（数字列 1-15）
    week_cols = [headers[w] for w in range(1, 16)]

    result = {}

    # 从第2行开始遍历数据
    for row in ws.iter_rows(min_row=2, values_only=True):
        # 读取值
        stream = str(row[col_stream - 1]) if row[col_stream - 1] is not None else ""
        code = str(row[col_code - 1]) if row[col_code - 1] is not None else ""
        first = str(row[col_first - 1]) if row[col_first - 1] is not None else ""
        second = str(row[col_second - 1]) if row[col_second - 1] is not None else ""

        # 读取周1-15
        week_data = {}
        for i, w_col in enumerate(week_cols):
            val = row[w_col - 1]
            val = str(val) if val is not None else ""
            if val.lower() == "none":
                val = ""
            week_data[f"Week_{i+1}"] = val

        # 构建 key
        stream = stream.replace("Y3_", "").replace("Telecom_", "Tele-").replace("IoT_G2", "IoT2").replace("IoT_G1", "IoT1")
        stream = stream.replace("TelecomG1", "Tele-G1").replace("TelecomG2", "Tele-G2")
        key = f"{stream}_{code}"

        # 存入字典
        result[key] = {
            "first_half": first,
            "second_half": second,
            "week_1_to_15": week_data
        }

    wb.close()
    return result
```

```python
REGION_LIST = [
    "B66:G76",
    "B78:G88",
    "B90:G100",
    "B102:G112",

    "B116:G126",
    "B128:G138",
    "B140:G150",
    "B152:G162",
]
# =================================================================

# 打开文件时 ⚠️ 加上 data_only=True ！！！（关键）
wb = load_workbook(EXCEL_FILE, data_only=True)
ws = wb["PDP"]

# 解析单元格坐标
def parse_cell(cell):
    col_str = re.findall(r'[A-Za-z]+', cell)[0]
    row_str = re.findall(r'\d+', cell)[0]
    return int(row_str), column_index_from_string(col_str)

# 解析区域
def parse_region(region_str: str):
    return region_str.replace(" ", "").split(":")

# 最终总字典
final_result = {}

# 批量处理所有区域
for region in REGION_LIST:    
    top_left, bottom_right = parse_region(region)
    start_row, start_col = parse_cell(top_left)
    end_row, end_col     = parse_cell(bottom_right)

    # 日期行（每个区域的第一行）
    date_row = start_row
    date_map = {}

    # ========== 【核心修复】读取真实日期值 ==========
    for col in range(start_col + 1, end_col + 1):
        cell = ws.cell(row=date_row, column=col)
        val = cell.value  # data_only=True 会拿到计算后的值        
        date_map[col] = val

    # 遍历数据行
    week_val = str(ws.cell(row=start_row, column=start_col - 1).value or "1").strip()
    for row in range(start_row + 1, end_row + 1):
        time_val = str(ws.cell(row=row, column=start_col).value or "").strip()

        week_val = week_val.replace("Week", "").replace(" ", "").replace("week", "")

        for col in range(start_col + 1, end_col + 1):
            cell = ws.cell(row=row, column=col)
            cell_name = cell.coordinate

            final_result[cell_name] = {
                "Date": date_map[col],
                "Time": time_val,
                "Week": week_val
            }

# 输出结果
for cell, info in final_result.items():
    print(f"{cell}: {info}")

print(f"\n🎉 总计提取：{len(final_result)} 个单元格")
```

# 排序

```python
def to_int(value, default=10**9):
    try:
        return int(str(value).strip())
    except Exception:
        return default


def day_sort_key(day: str) -> int:
    order = {
        "Monday": 1,
        "Tuesday": 2,
        "Wednesday": 3,
        "Thursday": 4,
        "Friday": 5,
        "Saturday": 6,
        "Sunday": 7,
    }
    return order.get(str(day).strip(), 99)


def time_sort_key(time_str: str) -> tuple:
    text = str(time_str or "").strip()
    m = re.match(r"^(\d{1,2}):(\d{2})", text)
    if not m:
        return (99, 99)
    return (int(m.group(1)), int(m.group(2)))


def room_sort_key(room: str) -> tuple:
    """
    Room 排序：
    1) 先排 3- 开头
    2) 再排 4- 开头
    3) '-' 后面的数字从大到小
    """
    text = str(room or "").strip()

    if text.startswith("3-"):
        prefix_rank = 0
    elif text.startswith("4-"):
        prefix_rank = 1
    else:
        prefix_rank = 2

    suffix_num = -1
    parts = text.split("-", 1)
    if len(parts) == 2:
        nums = re.findall(r"\d+", parts[1])
        if nums:
            suffix_num = int(nums[0])

    # 后缀降序 -> 用负号
    return (prefix_rank, -suffix_num, text)


def total_sort_key(item: dict) -> tuple:
    return (
        to_int(item.get("Week")),
        day_sort_key(item.get("Day")),
        time_sort_key(item.get("Time")),
        room_sort_key(item.get("Room")),
    )


total = sorted(total, key=total_sort_key)
```
