# Photo Upload Validation
核验 `temp.xlsx` 中每条记录在 OSS 上是否有对应的照片 `{month}.{day}_{start_time}_{room}.jpg`

```python
import pandas as pd
import hashlib
from IPython.display import display

# 加载 Excel
new_data = pd.read_excel("/root/BUPT-IS-CheckAttendApp/server/2026-03-13-1735-export.xlsx").to_dict(orient="records")
old_data = pd.read_excel("/root/BUPT-IS-CheckAttendApp/server/temp.xlsx").to_dict(orient="records")

for index, item in enumerate(old_data):
    hash_str = f"{item['Date']}{item['Time']}{item['Room']}"
    md5 = hashlib.md5(hash_str.encode('utf-8'))
    hash_value = md5.hexdigest()
    item["ID"] = hash_value

old_data = {item["ID"]: item for item in old_data}
```

```python
for item in new_data:
    info = old_data.get(item["ID"], {})
    
    if info:
        item["Student number in classroom"] = info["Student Num In Classroom"]
        item["Percent"] = f"{int(info['Percent (%)'] * 100)}"
        item["By"] = info["Filled By"]
```

```python
new_df = pd.DataFrame.from_records(new_data, index=None)
new_df.head()
```

```python
new_df.to_excel("/root/BUPT-IS-CheckAttendApp/server/2026-03-13-1833-export.xlsx", index=False)
```
